# Parte1

```python
email = "studente@esempio.it"
password = "Python123"
input_email = input("Inserisci email: ")#input email dell'utente
input_password = input("inserisci una password")# input password dell'utente

if email != input_email: #Se email inserita è diversa da quella richiesta
  print("Email non riconosciuta")
elif password != input_password:#Se la password inserita è diversa da quella richiesta
  print("Password errata")
else: #Altrimeti se email e password corrispondono
  print("Accesso consentito")

```

# Parte2

```python
input_number = int(input("Inserisci un numero intero:")) #input numerico
if input_number > 0 and input_number % 2 == 0:#Se il numero input è maggiore di zero e pari
  print("Numero positivo e pari")
elif input_number > 0 and input_number % 2 != 0:#Se il numero input è maggiore di zero e dispri
  print("Numero positivo e dispari ")
elif input_number == 0:#Se il numero input è zero
  print("Numero zero")
else: # altrimenti se non rispetta le due condizione è negativo
  print("Numero negativo")
```
