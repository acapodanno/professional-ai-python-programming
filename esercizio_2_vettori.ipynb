```python
class Vector():
  def __init__(self, sequence):
    self._sequence = sequence

  def __getitem__(self, index):
    assert(index < len(self._sequence)),'Indice non esistente'
    return self._sequence[index]

  def __len__(self):
    return len(self._sequence)
  def to_string(self):
    return str(self._sequence)
  def add(self, vector):
    if len(self._sequence) != len(vector._sequence):
      return None
    return Vector([x + y for x, y in zip(self._sequence, vector._sequence)])
  def sub(self, vector):
    if len(self._sequence) != len(vector._sequence):
      return None
    return Vector([x - y for x, y in zip(self._sequence, vector._sequence)])
  def prod(self, vector):
    if len(self._sequence) != len(vector._sequence):
      return None
    return Vector([x * y for x, y in zip(self._sequence, vector._sequence)])
  def equals(self, vector):
    return self._sequence == vector._sequence
```

```python
vector = Vector([1,2,3])
print(f"{vector.to_string()}")
try:
  print(f"{vector[5]}")
except Exception as e:
  print(e)

vector_2 = Vector([4,5,6])
print(f"{vector_2.to_string()}")
vector_3 = vector.add(vector_2)
print(f"{vector_3.to_string()}")
vector_4 = vector.sub(vector_2)
print(f"{vector_4.to_string()}")
vector_5 = vector.prod(vector_2)
print(f"{vector_5.to_string()}")

vector_6 = Vector([1,2,4])
print(f"{vector.equals(vector_6)}")
vector_7 = Vector([1,2,4,3])
print(f"{vector.add(vector_6)}")
```
