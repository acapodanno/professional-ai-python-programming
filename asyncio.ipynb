```python
import time
import asyncio

async def blocking_io(id,delay):
  print(f'Starting slow operation {id}')
  await asyncio.sleep(delay)
  print(f'Ending slow operation {id}')
  return f'Result from task {id}'
```

```python
async def main():
  start = time.time()
  tasks = [blocking_io(1,1),blocking_io(2,1),blocking_io(3,2)]
  for coroutine in await asyncio.gather(*tasks): # Corrected line: unpacking the list
    result = coroutine
    print(result)
  end = time.time()
  print(f'Time taken {end-start}')

async def main_2():
  start = time.time()
  tasks = [blocking_io(1,1),blocking_io(2,1),blocking_io(3,2)]
  for coroutine in asyncio.as_completed(tasks):
    result = await coroutine
    print(result)

  end = time.time()
  print(f'Time taken {end-start}')

await main()

await main_2()
```
