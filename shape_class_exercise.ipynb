```python
class Shape :
  def __init__(self, r):
    self._r=r

  def perimeter(self):
    """ Funzione per calcolare il perimetro"""
    return 2 * 3.14 * self._r
  def area(self):
    """ Funzione per calcolare il areo"""
    return 3.14 * (self._r * self._r)

```

```python
class Shape :
  def __init__(self, r):
    self._r=r

  def perimeter(self):
    return 2 * 3.14 * self._r
  def area(self):
    return 3.14 * (self._r * self._r)

shape = Shape(10)
print(f"{shape.area()}")
print(f"{shape.perimeter()}")
```
