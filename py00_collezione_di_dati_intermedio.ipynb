# List


```python
lista = ["Alessandro","Luca","Domenico","Roberto","Luca"]
```

```python
position = int(input(f"Inserisci la posizione da stampare tra 1 e {len(lista)}: "))
print(f"Nome da stampare è: {lista[position - 1]}")
```

```python
name = input("Digita un nome: ")
print(f"è presente: {name in lista}")
```

```python
lista.count(name)
print(f"Il nome è presente {lista.count(name)} volte!")
```

```python
new_name = input("Inserisci un nuovo nome:")
position = int(input(f"Inserisci la posizione:"))
lista.insert(position,name)
```

```python
remove_name = input("Inserisci il nome da rimuovere: ")
if remove_name in lista:
  lista.remove(remove_name)
else:
  print("Non è stato rimosso nessun elemento!")
remove_index = int(input(f"Inserisci la posizione da rimuovere tra 1 e {len(lista)}:"))
if remove_index < len(lista):
  print(f"è stato rimosso il nome {lista.pop(remove_index - 1)}")
else:
  print("Non è stato rimosso nessun elemento!")
```

# Set

```python
set = {'rosso', 'verde', 'blu', 'giallo', 'rosso', 'verde'}
print(f"Il set è il seguente: {set}")
set.add('nero')
print(f"Il set è il seguente: {set}")
set.discard("nero")
print(f"Il set è il seguente: {set}")
```

# Dizionario

```python
studenti = {"Marco":18,"Filippo":22,"Alessandro":30}
print(f"La lista è la seguente {studenti}")
studenti["Marco"] = 21
print(f"Gli studenti dopo la modifica: {studenti}")
studenti.update({"Marco":22})
print(f"Gli studenti dopo la modifica: {studenti}")
studenti.pop("Marco")
print(f"Gli studenti dopo la rimozione: {studenti}")

print(f"Le chiavi della lista sono: {studenti.keys()}")
print(f"I valori della lista sono: {studenti.values()}")
print(f"Gli studenti sono: {len(studenti)}")

```
