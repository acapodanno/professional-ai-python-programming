```python
class Calcolatrice:
  def __init__(self, numero1, numero2):
    self._numero1 = numero1
    self._numero2 = numero2

  def result(self,op):
    assert(op in ('+','-','*','/')), "Non esiste nessuna operazione disponibile"
    switcher = {
      '+': self._numero1 + self._numero2,
      '-': self._numero1 - self._numero2,
      '*': self._numero1 * self._numero2,
      '-': self._numero1 / self._numero2 #divisione per zero
    }
    return switcher.get(op)
```

```python
def input_numbers():
  return_number = False
  number = 0
  while not return_number:
    try:
      number = int(input("Inserisci il primo numero: "))
      return_number = True
    except ValueError:
      print("Non è un numero valido")
      return_number = False
  return number

calculate_again = True
while calculate_again:
  try:
    print(f"{Calcolatrice(input_numbers(),input_numbers()).result(input("Inserisci l'operazione:"))}")
    calculate_again = False
  except AssertionError as e:
    print(e)
```
