```python
# ho visto questa function su stackoverflow
def safe_cast(val, to_type, default=None):
    try:
        return to_type(val)
    except (ValueError, TypeError):
        return default



var_name = safe_cast(input("Insert your name: "),str,"")
var_age = safe_cast(input("Insert your age: "),int,0)
var_height = safe_cast(input("Insert your height:"),float,0.0)
var_born = 2024 - var_age
print(f"Your name is {var_name}, your age born is {var_born}, your height is {var_height:.2f}")
print(f"You are height {int(var_height*100)}cm")
```
